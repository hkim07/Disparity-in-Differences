```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Figure 1

```python
plt.figure(figsize=(5,4), dpi=300)

plt.subplot(221)
bb_disp_info = pd.read_csv("./outputs/journals_disp_info_by_th.csv")
N_T = bb_disp_info["n_nodes"].max()
E_T = bb_disp_info["n_edges"].max()
bb_disp_info["n_nodes"] = bb_disp_info["n_nodes"]/N_T
bb_disp_info["n_edges"] = bb_disp_info["n_edges"]/E_T
bb_disp_in_diffs_info = pd.read_csv("./outputs/journals_disp_in_diffs_info_by_th.csv")
bb_disp_in_diffs_info["n_nodes"] = bb_disp_in_diffs_info["n_nodes"]/N_T
bb_disp_in_diffs_info["n_edges"] = bb_disp_in_diffs_info["n_edges"]/E_T

plt.plot(bb_disp_info["n_edges"], bb_disp_info["n_nodes"], "--", label="Disparity Filter", linewidth=1.2, alpha=0.8)
plt.plot(bb_disp_in_diffs_info["n_edges"], bb_disp_in_diffs_info["n_nodes"], label="Disparity-in-Differences", linewidth=1.2, alpha=0.8)
plt.xlim(-0.02, 1.02)
plt.ylim(-0.02, 1.02)

plt.xlabel(r"$E/E_{T}$")
plt.ylabel(r"$N/N_{T}$")
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.legend(loc=4, fontsize=7.5, frameon=False)
plt.title("Journal Citation Network", fontsize=10)


plt.subplot(222)
bb_disp_info = pd.read_csv("./outputs/airports_disp_info_by_th.csv")
N_T = bb_disp_info["n_nodes"].max()
E_T = bb_disp_info["n_edges"].max()
bb_disp_info["n_nodes"] = bb_disp_info["n_nodes"]/N_T
bb_disp_info["n_edges"] = bb_disp_info["n_edges"]/E_T

bb_disp_in_diffs_info = pd.read_csv("./outputs/airports_disp_in_diffs_info_by_th.csv")
bb_disp_in_diffs_info["n_nodes"] = bb_disp_in_diffs_info["n_nodes"]/N_T
bb_disp_in_diffs_info["n_edges"] = bb_disp_in_diffs_info["n_edges"]/E_T

plt.plot(bb_disp_info["n_edges"], bb_disp_info["n_nodes"], "--", label="Disparity Filter", linewidth=1.2, alpha=0.8)
plt.plot(bb_disp_in_diffs_info["n_edges"], bb_disp_in_diffs_info["n_nodes"], label="Disparity-in-Differences", linewidth=1.2, alpha=0.8)
plt.xlim(-0.02, 1.02)
plt.ylim(-0.02, 1.02)

plt.xlabel(r"$E/E_{T}$")
plt.ylabel(r"$N/N_{T}$")
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.legend(loc=4, fontsize=7.5, frameon=False)
plt.title("U.S. Airport Network", fontsize=10)

plt.subplot(223)
bb_disp_info = pd.read_csv("./outputs/enron_disp_info_by_th.csv")
N_T = bb_disp_info["n_nodes"].max()
E_T = bb_disp_info["n_edges"].max()
bb_disp_info["n_nodes"] = bb_disp_info["n_nodes"]/N_T
bb_disp_info["n_edges"] = bb_disp_info["n_edges"]/E_T

bb_disp_in_diffs_info = pd.read_csv("./outputs/enron_disp_in_diffs_info_by_th.csv")
bb_disp_in_diffs_info["n_nodes"] = bb_disp_in_diffs_info["n_nodes"]/N_T
bb_disp_in_diffs_info["n_edges"] = bb_disp_in_diffs_info["n_edges"]/E_T

plt.plot(bb_disp_info["n_edges"], bb_disp_info["n_nodes"], "--", label="Disparity Filter", linewidth=1.2, alpha=0.8)
plt.plot(bb_disp_in_diffs_info["n_edges"], bb_disp_in_diffs_info["n_nodes"], label="Disparity-in-Differences", linewidth=1.2, alpha=0.8)
plt.xlim(-0.02, 1.02)
plt.ylim(-0.02, 1.02)

plt.xlabel(r"$E/E_{T}$")
plt.ylabel(r"$N/N_{T}$")
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.legend(loc=4, fontsize=7.5, frameon=False)
plt.title("Enron Email Network", fontsize=10)

plt.subplot(224)
bb_disp_info = pd.read_csv("./outputs/trade_disp_info_by_th.csv")
N_T = bb_disp_info["n_nodes"].max()
E_T = bb_disp_info["n_edges"].max()
bb_disp_info["n_nodes"] = bb_disp_info["n_nodes"]/N_T
bb_disp_info["n_edges"] = bb_disp_info["n_edges"]/E_T

bb_disp_in_diffs_info = pd.read_csv("./outputs/trade_disp_in_diffs_info_by_th.csv")
bb_disp_in_diffs_info["n_nodes"] = bb_disp_in_diffs_info["n_nodes"]/N_T
bb_disp_in_diffs_info["n_edges"] = bb_disp_in_diffs_info["n_edges"]/E_T

plt.plot(bb_disp_info["n_edges"], bb_disp_info["n_nodes"], "--", label="Disparity Filter", linewidth=1.2, alpha=0.8)
plt.plot(bb_disp_in_diffs_info["n_edges"], bb_disp_in_diffs_info["n_nodes"], label="Disparity-in-Differences", linewidth=1.2, alpha=0.8)
plt.xlim(-0.02, 1.02)
plt.ylim(-0.02, 1.02)

plt.xlabel(r"$E/E_{T}$")
plt.ylabel(r"$N/N_{T}$")
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=8)
plt.legend(loc=4, fontsize=7.5, frameon=False)
plt.title("World Trade Network", fontsize=10)


plt.tight_layout(h_pad=1.2, w_pad=0.3)
plt.savefig("./figs/fig1.pdf", bbox_inches="tight")
plt.savefig("./figs/fig1.png", bbox_inches="tight")
```

### Figure 2

```python
abdc_eval = pd.read_csv("./outputs/abdc_eval_agg.csv")
airport_eval = pd.read_csv("./outputs/airports_eval_agg.csv")
airport_eval["region"] = airport_eval["region"].str.replace("Northwest Mountain", "Northwest\nMountain")
airport_eval["region"] = airport_eval["region"].str.replace("Western-Pacific", "Western-\nPacific")
airport_eval["region"] = airport_eval["region"].str.replace("Great Lakes", "Great\nLakes")
airport_eval["region"] = airport_eval["region"].str.replace("New England", "New\nEngland")

plt.figure(figsize=(5,5), dpi=300)

plt.subplot(211)

abdc_eval_agg_dict = {
    "Disparity Filter": abdc_eval["misaligned_rate_disp"],
    "Disparity-in-Differences": abdc_eval["misaligned_rate_disp_in_diffs"]
}

x = np.arange(len(abdc_eval))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

for attribute, measurement in abdc_eval_agg_dict.items():
    offset = width * multiplier
    rects = plt.bar(x + offset, measurement, width, label=attribute, alpha=0.8)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

plt.xlabel('Field of Research', fontsize=10)
plt.ylabel('Misaligned Proportion', fontsize=10)
plt.xticks(x+width/2, abdc_eval["FoR"], rotation=90, fontsize=8)
plt.yticks([0.0, 0.1, 0.2, 0.3], ["0.00", "0.10", "0.20", "0.30"], fontsize=8)
plt.legend(loc=0, ncols=2, fontsize=7.5, frameon=False)
plt.ylim(0, 0.33)
plt.title("ABDC Indexed Journals", fontsize=10)

plt.subplot(212)

airport_eval_agg_dict = {
    "Disparity Filter": airport_eval["misaligned_rate_disp"],
    "Disparity-in-Differences": airport_eval["misaligned_rate_disp_in_diffs"]
}

x = np.arange(len(airport_eval))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

for attribute, measurement in airport_eval_agg_dict.items():
    offset = width * multiplier
    rects = plt.bar(x + offset, measurement, width, label=attribute, alpha=0.8)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

plt.xlabel('Region', fontsize=10)
plt.ylabel('Misaligned Proportion', fontsize=10)
plt.xticks(x+width/2, airport_eval["region"], rotation=90, fontsize=8)
plt.yticks(fontsize=8)
#plt.yticks([0.0, 0.02, 0.04, 0.06, 0.08], ["0.00", "0.02", "0.04", "0.06", "0.08"], fontsize=8)
plt.legend(loc=0, ncols=2, fontsize=7.5, frameon=False)
plt.ylim(0, 0.11)
plt.title("U.S. Hub Airports", fontsize=10)

plt.tight_layout(pad=1.2)
plt.savefig("./figs/fig2.pdf", bbox_inches="tight")
plt.savefig("./figs/fig2.png", bbox_inches="tight")
```
